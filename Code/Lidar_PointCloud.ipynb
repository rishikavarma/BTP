```python
import csv
import numpy as np
import matplotlib.pyplot as plt
file = open( 'lidar_pcl/bedroom1_data/sweep-data/2022-04-10 10_58_00_578226.csv' ) #input file path
csvreader = csv.reader( file )
data = []
x = []
y = []
for row in csvreader:
    signal = row[ 2 ]
    signal = signal[ 1 : -2 ]
    arr = signal.split( '), Sample(')
    arr[ 0 ] = arr[ 0 ].split( 'Sample(' )[ 1 ]
    row_data = []
    for r in arr:
        temp = r.split( ',' )
        p = ( int( temp[ 0 ].split( '=' )[ 1 ]  ) / 1000, int( temp[ 1 ].split( '=' )[ 1 ] ), int( temp[ 2 ].split( '=' )[ 1 ] ) )
        xi = p[ 1 ] * np.sin( p[ 0 ] * np.pi / 180 )
        yi = p[ 1 ] * np.cos( p[ 0 ] * np.pi / 180 )
        x.append( xi )
        y.append( yi )
        row_data.append( p )
    data.append( row_data )
# print( data )
plt.scatter( x, y )
plt.show()
plt.figure(figsize=(10,10))
plt.scatter( x, y, s=0.5)
axes=plt.gca()
axes.set_aspect('equal')
plt.show()
# print( x )
```
